# Transformación matricial de un <u>vector</u> 

Una matriz $A$ de $m \times n$, define la **transformación matricial de un vector**.

$T_A : \mathbb{R}^n \rightarrow \mathbb{R}^m$

Dada por 

$T_A (\vec{x}) = A \vec{x} $

# Transformación matricial de una <u>matriz</u>

Cuando hay que aplicar la misma transformación a varios vectores, como se muestra a continuación: 
* $T_A(\vec{v}_1)=\vec{u}_1$
* $T_A(\vec{v}_2)=\vec{u}_2$
* $\vdots$
* $T_A(\vec{v}_n)=\vec{u}_n$

Se puede simplificar la notación colocando los vectores en matrices.
* $B=[ \matrix{ \vec{v}_1, \vec{v}_2, \cdots, \vec{v}_n}]$
* $C=[ \matrix{ \vec{u}_1, \vec{u}_2, \cdots, \vec{u}_n}]$

Entonces en vez de escribir muchas transformaciones matriciales de vectores podemos escribir sólo una transformación matricial de matriz. 

$T_A^*(B)=C$

Usaremos el asterisco para hacer énfasis que la transformación matricial es sobre una matriz y no sobre un vector.

Debido a la definición de multiplicación matricial

$A [ \matrix{ \vec{v}_1, \vec{v}_2, \cdots, \vec{v}_n}]=[ \matrix{ A\vec{v}_1, A\vec{v}_2, \cdots, A\vec{v}_n}]$

Se puede definir de manera general la **transformación matricial de una matriz** $B$ de la siguiente forma.

$T_A^*(B)  = A B $



# Linealidad 

Es de resaltar en este curso la importancia de:
* la suma de vectores y
* la multiplicación de escalar por vector. 

Recordará que estas operaciones juegan un papel fundamental en la definición de un subespacio de $\mathbb{R}^n$. Ahora vuelven a tomar un papel estelar en la definición de linealidad.






Una transformación $T_A : \mathbb{R}^n \rightarrow \mathbb{R}^m$ es **lineal** cuando cumple: 
* (AT) $T(\vec{u}_1 + \vec{u}_2)=T(\vec{u}_1) + T(\vec{u}_2)$
* (MT) $T(c_1\vec{u}_1)=c_1T(\vec{u}_1)$

para todo $\vec{u}_1,\vec{u}_2 \in \mathbb{R}^n$ y para todo $c_1, c_2 \in \mathbb{R}$

(AT) y (MT) equivalen a la siguiente igualdad 

* $T(c_1\vec{u}_1 + c_2\vec{u}_2)=c_1T(\vec{u}_1) + c_2T(\vec{u}_2)$

Utilizando inducción matemática se puede mostrar la equivalencia de (AT) y (MT) con la siguiente igualdad 

* $T(c_1\vec{u}_1 + c_2\vec{u}_2 + \cdots + c_n\vec{u}_n)=c_1T(\vec{u}_1) + c_2T(\vec{u}_2) + \cdots + c_nT(\vec{u}_n)$

para todo $\vec{u}_1,\vec{u}_2, \ldots , \vec{u}_n \in \mathbb{R}^n$ y para todo $c_1, c_2, \ldots, c_n \in \mathbb{R}$

Esto permite concluir que una transformación es lineal si 
la transformación de cualquier combinación lineal de vectores es la misma combinación lineal de la transformación de los vectores.


# Linealidad y Transformación matricial
$T:\mathbb{R}^n \rightarrow \mathbb{R}^m$ es una transformación lineal **si y sólo si** para todo $v \in \mathbb{R}^n$ se cumple que $T(v)=Av$. Donde $A$ es la matriz de $m \times n$  dada por 
 $$A=T^*(I_n)= \begin{bmatrix}
  T \left( \begin{matrix} 1 \\ 0 \\ \vdots \\ 0 \end{matrix} \right) & 
  T \left( \begin{matrix} 0 \\ 1 \\ \vdots \\ 0 \end{matrix} \right) &
  \cdots &
  T \left( \begin{matrix} 0 \\ 0 \\ \vdots \\ 1 \end{matrix} \right) 
 \end{bmatrix} $$ 
 
 
 ## Prueba
 Como la anterior afirmación tiene "**si y sólo si**" entonces hay que hacer dos demostraciones. Primero mostraremos que **si** $T:\mathbb{R}^n \rightarrow \mathbb{R}^m$ es una transformación lineal **entonces** para todo $\vec{v} \in \mathbb{R}^n$ se cumple que $T(\vec{v})=A\vec{v}$.

$\begin{align}
T(\vec{v})&=T(I_n \vec{v})\\
&=T\left(
\left[ \begin{matrix} 1 &0 & \cdots & 0\\ 0 &1 & \cdots & 0 \\ \vdots &\vdots & \ddots & \vdots \\ 0 &0 & \cdots & 1 \end{matrix} \right] 
\left( \begin{matrix} c_1 \\ c_2 \\ \vdots \\ c_n \end{matrix} \right) 
\right)\\
&=T\left(
c_1\left( \begin{matrix} 1 \\ 0 \\ \vdots \\ 0 \end{matrix} \right) 
+ \cdots 
+ c_n\left( \begin{matrix} 0 \\ 0 \\ \vdots \\ 1 \end{matrix} \right) 
\right)\\
&=c_1T\left( \begin{matrix} 1 \\ 0 \\ \vdots \\ 0 \end{matrix} \right) 
+ \cdots 
+ c_nT\left( \begin{matrix} 0 \\ 0 \\ \vdots \\ 1 \end{matrix} \right)\\
&=\begin{bmatrix}
  T \left( \begin{matrix} 1 \\ 0 \\ \vdots \\ 0 \end{matrix} \right) & 
  T \left( \begin{matrix} 0 \\ 1 \\ \vdots \\ 0 \end{matrix} \right) &
  \cdots &
  T \left( \begin{matrix} 0 \\ 0 \\ \vdots \\ 1 \end{matrix} \right) 
 \end{bmatrix} 
 \left( \begin{matrix} c_1 \\ c_2 \\ \vdots \\ c_n \end{matrix} \right)\\
&=A\vec{v}
\end{align}$

 
 
 

Ahora mostraremos que  $T:\mathbb{R}^n \rightarrow \mathbb{R}^m$ es una transformación lineal **si** para todo $v \in \mathbb{R}^n$ se cumple que $T(\vec{v})=A\vec{v}$. 

Para esto hay que probar que $T(c_1\vec{u}_1 + c_2\vec{u}_2)=c_1T(\vec{u}_1) + c_2T(\vec{u}_2)$.

$\begin{align}
T(c_1\vec{u}_1 + c_2\vec{u}_2)&=A(c_1\vec{u}_1 + c_2\vec{u}_2)\\
&=c_1A\vec{u}_1 + c_2A\vec{u}_2\\
&=c_1T(\vec{u}_1) + c_2T(\vec{u}_2)
\end{align}$


# Ejemplo de linealidad
Demuestre que $T \left( \begin{matrix} x \\ y \\ z \end{matrix} \right) = \begin{pmatrix} 3x - z \\ y  \end{pmatrix} $ es una transformación lineal y encuentre la respectiva matriz.


## Solución
Primero encontraremos la matriz $A=T^*(I)$ y luego comprobaremos que $A\begin{pmatrix} x \\ y \\ z \end{pmatrix}=T \left( \begin{matrix} x \\ y \\ z \end{matrix} \right)$

Vamos a encontrar la matriz $A$.

\begin{align}
A &=T^*\left(\begin{bmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{bmatrix}\right)\\
  &=\begin{bmatrix}
  T \left( \begin{matrix} 1 \\ 0 \\ 0 \end{matrix} \right) & 
  T \left( \begin{matrix} 0 \\ 1 \\ 0 \end{matrix} \right) &
  T \left( \begin{matrix} 0 \\ 0 \\ 1 \end{matrix} \right) 
 \end{bmatrix}\\
 &=\begin{bmatrix}
 3 & 0 & -1 \\ 
 0 & 1 &  0
 \end{bmatrix}
\end{align}

Ahora probamos que $A\begin{pmatrix} x \\ y \\ z \end{pmatrix}=T \left( \begin{matrix} x \\ y \\ z \end{matrix} \right) $

$A\begin{pmatrix} x \\ y \\ z \end{pmatrix}=\begin{bmatrix}
 3 & 0 & -1 \\ 
 0 & 1 &  0
 \end{bmatrix}
\begin{pmatrix} x \\ y \\ z \end{pmatrix}
=\begin{pmatrix} 3x - z \\ y  \end{pmatrix} 
=T \left( \begin{matrix} x \\ y \\ z \end{matrix} \right)$

Lo cual prueba que $T$ sí es lineal.




# Ejemplo de no linealidad
Demuestre que $T \left( \begin{matrix} x \\ y \\ z \end{matrix} \right) = \begin{pmatrix} 3 - z \\ y  \end{pmatrix} $ no es una transformación lineal.


## Solución
Primero encontraremos la matriz $A=T^*(I)$ y luego verificamos si $A\begin{pmatrix} x \\ y \\ z \end{pmatrix}=T \left( \begin{matrix} x \\ y \\ z \end{matrix} \right)$

Vamos a encontrar la matriz $A$.

\begin{align}
A &=T^*\left(\begin{bmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{bmatrix}\right)\\
  &=\begin{bmatrix}
  T \left( \begin{matrix} 1 \\ 0 \\ 0 \end{matrix} \right) & 
  T \left( \begin{matrix} 0 \\ 1 \\ 0 \end{matrix} \right) &
  T \left( \begin{matrix} 0 \\ 0 \\ 1 \end{matrix} \right) 
 \end{bmatrix}\\
 &=\begin{bmatrix}
 3 & 3 & 2 \\ 
 0 & 1 &  0
 \end{bmatrix}
\end{align}

Ahora probamos que $A\begin{pmatrix} x \\ y \\ z \end{pmatrix}=T \left( \begin{matrix} x \\ y \\ z \end{matrix} \right) $

$A\begin{pmatrix} x \\ y \\ z \end{pmatrix}=\begin{bmatrix}
 3 & 3 & 2 \\ 
 0 & 1 &  0
 \end{bmatrix}
\begin{pmatrix} x \\ y \\ z \end{pmatrix}
=\begin{pmatrix} 3x + 3y + 2z \\ y  \end{pmatrix} 
\neq
\begin{pmatrix} 3 - z \\ y  \end{pmatrix}=
 T \left( \begin{matrix} x \\ y \\ z \end{matrix} \right)
$

Lo cual prueba que $T$ no es lineal.

# Obtener una transformación a partir de la transformación de una base

image.png


Sea $B=[ \matrix{ \vec{v}_1, \vec{v}_2, \cdots, \vec{v}_n}]$  una matriz invertible de $n \times n$.

Además, se conoce que la transformación $T_A$ cumple lo siguiente para alguna matriz $A$:
* $T_A(\vec{v}_1)=\vec{u}_1$
* $T_A(\vec{v}_2)=\vec{u}_2$
* $\vdots$
* $T_A(\vec{v}_n)=\vec{u}_n$

Encuentre la matriz $A$.

Solución.
Si $C=[ \matrix{ \vec{u}_1, \vec{u}_2, \cdots, \vec{u}_n}]$

Entonces podemos reescribir todas las transformaciones matriciales de cada <u>vector</u> 
* $T_A(\vec{v}_1)=\vec{u}_1$
* $T_A(\vec{v}_2)=\vec{u}_2$
* $\vdots$
* $T_A(\vec{v}_n)=\vec{u}_n$

como una transformación matricial de una <u>matriz</u>

$T_A^*(B)=C$

De acuerdo con la definición dada al inicio, esto se puede escribir como 

$AB=C$

Como $B$ es una matriz invertible entonces podemos despejar la matriz $A$

$A=CB^{-1}$

## Ejemplo.
Si se sabe que 
* $T_A\left( \matrix{1\\4\\6} \right)=\left( \matrix{2\\2} \right)$
* $T_A\left( \matrix{1\\2\\1} \right)=\left( \matrix{2\\2} \right)$
* $T_A\left( \matrix{0\\1\\3} \right)=\left( \matrix{1\\3} \right)$

Encuentre la matriz $A$.

Solución.

Las matrices $B$ y $C$ se construyen a partir de los vectores dados.

```python
B=sp.Matrix([[1,1,0],
             [4,2,1],
             [6,1,3]])
B
```

```python
C=sp.Matrix([[2,2,1],
             [2,2,3]])
C
```

Ahora es se encuentra la inversa de la matriz $B$

```python
M=B.row_join(sp.eye(3))
M
```

```python
M[1,:] += -4*M[0,:]
M[2,:] += -6*M[0,:]
M
```

```python
M[2,:] += M[1,:]*-5/2
M[0,:] += M[1,:]/2
M
```

```python
M[0,:] -= M[2,:]
M[1,:] += -2*M[2,:]
M
```

```python
M[1,:] *= -sp.Rational(1,2)
M[2,:] *= 2
M
```

```python
B_inv=M[:,3:]
B_inv
```

Ahora verificamos que la inversa es correcta. Este paso es importante especialmente cuando los cálculos se realizan manualmente.

```python
B*B_inv  
```

Finalmente se calcula la matriz $A$

```python
A=C*B_inv
A
```

Ahora se comprueba que el resultado corresponda con lo solicitado.

```python
A*sp.Matrix([1,
             4,
             6])
```

```python
A*sp.Matrix([1,
             2,
             1])
```

```python
A*sp.Matrix([0,
             1,
             3])
```
