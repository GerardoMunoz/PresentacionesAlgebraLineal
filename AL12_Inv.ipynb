# Bases de $\mathbb{R}^m$

Si una matriz $A$ es una base de $\mathbb{R}^m$ entonces:

* $A$ es Linealmente independiente $\Rightarrow$ $A$ tiene un l-pivote en cada columna.
* $A$ genera todo $\mathbb{R}^m$ $\Rightarrow$ $A$ tiene un l-pivote en cada renglón.

La única forma para que una matriz tenga un l-pivote en cada renglón y en cada columna es que la matriz sea cuadrada y que los pivotes estén sobre la diagonal. 

```python
import sympy as sp
M=sp.Matrix([[2,0,5],[0,-2,0],[-1,1,2]])
M
```

```python
M.row_swap(0,2)
M
```

```python
M[2,:] += 2*M[0,:]
M
```

```python
M[2,:] += M[1,:]
M
```

```python
M[2,:] = M[2,:]/9
M
```

```python
M[0,:] -= 2*M[2,:]
M
```

```python
M[1,:] = M[1,:]/-2
M
```

```python
M[0,:] -= M[1,:]
M
```

```python
M[0,:] = -M[0,:]
M
```

Además, la forma escalón reducida de una base de $\mathbb{R}^m$ siempre es la matriz identidad.

# Base y sistemas de ecuaciones

Si $B$ es una base de $\mathbb{R}^m$ entonces el sistema de ecuaciones dado por 

$$B\vec{x}=\vec{b}$$

Tiene solución <u>única</u> para <u>todo</u> $\vec{b} \in \mathbb{R}^m$ 

* Tiene solución única porque cada columna de $b$ tiene l-pivote.

*  Para todo $\vec{b} \in \mathbb{R}^m$ porque cada renglón de $b$ tiene l-pivote.

¿Cuál es la solución única?
Si el sistema es homogéneo ($\vec{b}=\vec{0}$) la solución única es $\vec{x}=0$.

Vamos a ver que si el sistema $B\vec{x}=\vec{b}$ no es homogéneo podremos encontrar una matriz $A$ tal que $AB=I_n$. De esta forma se puede despejar $\vec{x}$.

$$B\vec{x}=\vec{b}$$
$$AB\vec{x}=A\vec{b}$$
$$I_n\vec{x}=A\vec{b}$$
$$\vec{x}=A\vec{b}$$
