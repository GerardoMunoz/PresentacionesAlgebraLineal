Para probar esto, es necesario probar los axiomas:
* (A1) $v_1, v_2 \in V \Rightarrow v_1 + v_2 \in V$ 
* (M1) $v_1 \in V \text{ y } k \in R \Rightarrow kv_1  \in V$ 

Comencemos con (M1)

Si 
$$ v_1=B \text{cos}(\chi + \beta)$$
entonces 
$$ kv_1=kB \text{cos}(\chi + \beta)$$
haciendo $A=kB$ y $\alpha=\beta$ vemos $kv_1 \in V$
y por lo tanto cumple (M1)

Al parecer sí es una función de la forma $A \text{cos}(\chi + \alpha)$. Pero ¿cuáles serán los valores de $A$ y de $\alpha$? La respuesta a esta pregunta la encontraremos usando la identidad trigonométrica 
\begin{aligned}
A \text{cos}(\chi + \alpha) 
&= \big(A\text{cos}(\alpha)\big)\text{cos}(\chi)-\big(A\text{sin}( \alpha)\big)\text{sin}(\chi)\\
&= A_x\text{cos}(\chi)-A_y\text{sin}(\chi)\\
\end{aligned}

Como $A$ y $\alpha$ son constantes, entonces $A_x=A\text{cos}(\alpha)$ y $A_y=A\text{sin}( \alpha)$ también son constantes que de ahora en adelante llamaremos $A_x$ y $A_y$. Incluso podemos representarlas como un vector $\vec{u}=\begin{pmatrix}A_x\\A_y\end{pmatrix} \in \mathbb{R}^2$. La representación geométrica de este vector es simplemente ¡espectacular!


Como $A, A_x, A_y$ forman un triángulo rectángulo, estos cumplen el teorema de Pitágoras y las funciones trigonométricas.

$$A_x^2 + A_y^2=A^2\big(\cos^2(\alpha)+\sin^2( \alpha)\big)=A^2$$
$$\frac{A_y}{A_x}=\frac{A\sin(\alpha)}{A\cos( \alpha)}=\tan(\alpha)$$

Esto quiere decir que una función de la forma $A \text{cos}(\chi + \alpha)$ puede ser representada por un vector $\vec{u}=\begin{pmatrix}A_x\\A_y\end{pmatrix} \in \mathbb{R}^2$. Esto se llama un fasor, ya que el ángulo de fase del vector corresponde al ángulo del vector $u$. 

Ahora viene lo más interesante. La suma de $$B \cos(\chi + \beta) + C \cos(\chi + \gamma)$$ usando fasores.

\begin{aligned}
&B \cos(\chi + \beta) + C \cos(\chi + \gamma)\\
&=B_x \cos(\chi) - B_y \sin(\chi) + C_x \cos(\chi) - C_y \sin(\chi)\\
&=(B_x +C_x)\cos(\chi) - (B_y + C_y) \sin(\chi) \\
&=A_x\cos(\chi) - A_y \sin(\chi) \\
\end{aligned}

En donde $A_x$ y $A_y$ escritos vectorialmente se pueden escribir así
$$\begin{pmatrix}A_x\\A_y\end{pmatrix}=\begin{pmatrix}B_x+C_x\\B_y+C_y\end{pmatrix}=\begin{pmatrix}B_x\\B_y\end{pmatrix}+\begin{pmatrix}C_x\\C_y\end{pmatrix}$$

Lo cual significa que ¡La suma de las funciones coseno corresponde con la suma de sus fasores! 

Entonces ya podemos encontrar $A$ y $\alpha$ para que 
$$A \text{cos}(\chi + \alpha)=0.7\cos(\chi-60°)+0.5\cos(\chi+30°)$$

# Base del espacio vectorial de las funciones coseno

A continuación, vamos a mostrar que el conjunto de todas las funciones coseno
$$V=\{A \text{cos}(\chi + \alpha) \mid A,\alpha\in \mathbb{R} \}$$
puede ser generado de manera única por tan sólo dos funciones
$$\{\cos(\chi), -\sin(\chi)\}$$
Primero vemos que cualquier función coseno $A \text{cos}(\chi + \alpha)$ puede ser escrita como combinación lineal de las funciones $\{\cos(\chi), -\sin(\chi)\}$. Esto se logra gracias a la identidad trigonométrica vista anteriormente
\begin{aligned}
A \text{cos}(\chi + \alpha) 
&= \big(A\text{cos}(\alpha)\big)\text{cos}(\chi)-\big(A\text{sin}( \alpha)\big)\text{sin}(\chi)\\
&= A_x\text{cos}(\chi)-A_y\text{sin}(\chi)\\
\end{aligned}
Esto demuestra que $\text{Gen}\{\cos(\chi), -\sin(\chi)\}=V$

Ahora miramos que los coeficientes $A_x, A_y$ son únicos. Esto se logra igualando a la función cero 
$$A_x\text{cos}(\chi)-A_y\text{sin}(\chi)=0(\chi)$$
Como esto se debe cumplir para todos los valores de $\chi$. En particular se debe cumplir para $\chi=0$ y $\chi=90°$.
* Si $\chi=0$ entonces $A_x\text{cos}(0)-A_y\text{sin}(0)=0$. Por lo tanto $A_x=0$.
* Si $\chi=90°$ entonces $A_x\text{cos}(90°)-A_y\text{sin}(90°)=0$. Por lo tanto $A_y=0$.

Lo cual demuestra que $\{\cos(\chi), -\sin(\chi)\}$ es L.I.

Como $\{\cos(\chi), -\sin(\chi)\}$ genera a $V$ y es L.I entonces es una base de $V$. Por lo tanto, la dimensión de $V$ es 2.
