# Transformación matricial de un <u>vector</u> 
# y transformación matricial de una <u>matriz</u>.

Una matriz $A$ de $m \times n$, define la transformación matricial de un vector.

$T_A : \mathbb{R}^n \rightarrow \mathbb{R}^m$

Dada por 

$T_A (\vec{x}) = A \vec{x} $

<!-- \left( \matrix{x_1\\ x_2 \\ \vdots \\x_n} \right) =A-->

De manera similar define la transformación matricial de una matriz $B=[v_1 \ \ v_2 \ \  \cdots \ \ v_k ] $ de tamaño $n \times k$, para algún $k \in \mathbb{N}$.

$\begin{align}
T_A^*(B)  &=[T_A(v_1) \ \ T_A(v_2) \ \  \cdots \ \ T_A(v_k) ] \\&=[Av_1 \ \ Av_2 \ \  \cdots \ \ Av_k ] \\&= A[v_1 \ \ v_2 \ \  \cdots \ \ v_k ] \\&= A B 
\end{align}$

La matriz cero transforma todo al origen, por eso no consideraremos este caso.

# Compresión o expansión

Se aumenta o se disminuye la escala en uno o en ambos ejes. Se obtiene al multiplicar un renglón por una constante.

$\left[ \matrix{2 & 0\\ 0 & 1} \right]
\left( \matrix{x_0\\ x_1} \right)=\left( \matrix{2x_0\\ x_1} \right)$


```python
A=sp.Matrix([[2,0],
             [0,1]])

transf226(casa,A*casa,-1,-2,3,2)
```

$\left[ \matrix{1 & 0\\ 0 & \frac{1}{2}} \right]
\left( \matrix{x_0\\ x_1} \right)=\left( \matrix{x_0\\ \frac{1}{2}x_1} \right)$

```python
A=sp.Matrix([[1,0],
             [0,1/2]])

transf226(casa,A*casa,x0=-2,x1=2)
```

# Intercambio de ejes

Se refleja con respecto a la recta $x_0 = x_1$. Se intercambian los renglones.

$\left[ \matrix{0 & 1\\ 1 & 0} \right]
\left( \matrix{x_0\\ x_1} \right)=\left( \matrix{x_1\\ x_0} \right)$



```python
A=sp.Matrix([[0,1 ],
             [1,0   ]])

transf226(casa,A*casa)
```

# Deslizamiento

El valor de un eje afecta el otro eje. Se obtiene al sumar un múltiplo de un renglón a otro.

$\left[ \matrix{1 & 0.5\\ 0 & 1} \right]
\left( \matrix{x_0\\ x_1} \right)=\left( \matrix{x_0 + 0.5x_1\\ x_1} \right)$



```python
A=sp.Matrix([[1, 0.5 ],
             [0, 1   ]])

transf226(casa,A*casa)
```



$\left[ \matrix{1 & 0\\ 0.5 & 1} \right]
\left( \matrix{x_0\\ x_1} \right)=\left( \matrix{x_0 \\ 0.5x_0 + x_1} \right)$



```python
A=sp.Matrix([[1, 0 ],
             [0.5, 1   ]])

transf226(casa,A*casa)
```
